```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px

ori = pd.read_csv("EPL_Soccer_MLR_LR.csv")

print(sns.heatmap(ori.corr("pearson", numeric_only = True), cmap='viridis'))
```

```python
sns.clustermap(ori.corr(), 
               annot = True,      
               cmap = 'RdYlBu_r',  
               vmin = -1, vmax = 1,
              )
```

```python
np.triu(np.ones_like(ori.corr()))

mask = np.triu(np.ones_like(ori.corr(), dtype=np.bool))

fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(ori.corr(), 
            mask=mask, 
            vmin=-1, 
            vmax = 1, 
            annot=True, 
            cmap="RdYlBu_r", 
            cbar = True)
ax.set_title('Wine Quality Correlation', pad = 15)
```
